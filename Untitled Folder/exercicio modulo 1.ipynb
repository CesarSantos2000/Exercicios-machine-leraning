```python
y = concate['StyleID']
x = concate.drop('StyleID', axis=1)
```

```python
def classificador (a, b): 
    from sklearn.linear_model import LogisticRegression
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.naive_bayes import GaussianNB
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import StratifiedKFold
    import numpy as np
    skfold  = StratifiedKFold(n_splits=3)
    x = a
    y = b
    normalizador = MinMaxScaler(feature_range = (0, 1))
    x_norm = normalizador.fit_transform(x)
    
    logistic = LogisticRegression()
    decision = DecisionTreeRegressor()
    gaussian = GaussianNB()
    knn = KNeighborsClassifier()
    
    resul_logistic = cross_val_score(logistic,x,y,cv = skfold)
    resul_decision = cross_val_score(decision,x,y,cv = skfold)
    resul_gaussian = cross_val_score(gaussian,x,y,cv = skfold)
    resul_knn = cross_val_score(knn,x_norm,y,cv = skfold)
    
    dic_modelos = {'Logistica': resul_logistic.mean(), 'Decisão': resul_decision.mean(),'Naive': resul_gaussian.mean(), 'KNN': resul_knn.mean()}
    melhor_modelo = max(dic_modelos, key=dic_modelos.get)
    print('Regressão Logistica ', resul_logistic.mean(), 'Decision Tree ', resul_decision.mean(), 'Naive Bayers ', resul_gaussian.mean(), 'KNN ', resul_knn.mean())
    print('Melhor modelo: ', melhor_modelo, 'Valor: ', dic_modelos[melhor_modelo])
```

```python
classificador(x, y)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV

normalizador = MinMaxScaler(feature_range = (0, 1))
x_norm = normalizador.fit_transform(x)

valoresk = np.array([1, 3, 5, 9, 13, 17, 21, 25])
calc_distanci = ['minkowski', 'chebyshev']
valoresp = np.array([1, 2, 3])
gridvalue = {'n_neighbors': valoresk, 'metric':calc_distanci, 'p':valoresp}

modelo = KNeighborsClassifier()

resultKNN = GridSearchCV(estimator=modelo, param_grid=gridvalue, cv=5, n_jobs=-1)
resultKNN.fit(x_norm,y)

print('Melhor acurácia', resultKNN.best_score_)
print('Melhor K', resultKNN.best_estimator_.n_neighbors)
print('Metodo distancia', resultKNN.best_estimator_.metric)
print('Melhor p', resultKNN.best_estimator_.p)
```
